# ใบงานที่ 1 เรียนรู้การสร้าง class ในภาษา C#

## ตอนที่ 1:  ทำความเข้าใจ classes

### Class 

ในภาษาโปรแกรมเชิงวัตถุ (Object Oriented Programming) คลาส (Class) คือพิมพ์เขียวหรือแม่แบบที่ใช้อธิบายลักษณะและพฤติกรรมของ อ็อบเจ็กต์ (Object) ซึ่งเป็นตัวแทนของสิ่งต่างๆ ในโลกแห่งความจริง เช่น คน สัตว์ สิ่งของ (tangible) หรือแนวคิด (intangible)



### คลาสใน C# ประกอบด้วยสมาชิกต่างๆ ดังนี้

ฟิลด์ (Fields): คือตัวแปรที่เก็บข้อมูลหรือสถานะของอ็อบเจ็กต์ เช่น ชื่อ อายุ สี
เมธอด (Methods): คือฟังก์ชันที่กำหนดพฤติกรรมหรือการกระทำของอ็อบเจ็กต์ เช่น เดิน วิ่ง กิน
คอนสตรักเตอร์ (Constructors): คือเมธอดพิเศษที่ใช้สร้างอ็อบเจ็กต์ของคลาส
พร็อพเพอร์ตี้ (Properties): คือวิธีที่เข้าถึงและแก้ไขข้อมูลของฟิลด์อย่างปลอดภัย


### หลักการสำคัญของ OOP ที่เกี่ยวข้องกับคลาส

#### การห่อหุ้ม (Encapsulation)
คือการซ่อนข้อมูลภายในของอ็อบเจ็กต์และให้เข้าถึงผ่านเมธอดหรือพร็อพเพอร์ตี้เท่านั้น เพื่อความปลอดภัยและป้องกันการเปลี่ยนแปลงข้อมูลโดยตรง

#### การสืบทอด (Inheritance)
คือการสร้างคลาสใหม่ที่สืบทอดคุณสมบัติและพฤติกรรมจากคลาสเดิม ทำให้โค้ดมีความยืดหยุ่นและลดการเขียนโค้ดซ้ำซ้อน

### การมีหลากหลายรูปแบบ (Polymorphism)
คือการที่อ็อบเจ็กต์หลายประเภทสามารถตอบสนองต่อเมธอดเดียวกันในรูปแบบที่แตกต่างกัน

### ประโยชน์ของคลาส


โดยทั่วไป การออกแบบคลาส จะออกแบบให้แต่ละคลาสมีวัตถุประสงค์เฉพาะอย่าง และไม่นำสิ่งที่ไม่เกี่ยวข้องใมาใส่ไว้ในคลาสโดยเด็ดขาด ทำให้เราสามารถเพ่งความสนใจไปในคลาสนั้นได้อย่างเต็มที่ 

การจัดกลุ่มข้อมูลและพฤติกรรมที่เกี่ยวข้องเข้าไว้ในคลาสเดียวกัน ช่วยให้เราแสดงสิ่งต่างๆ ในโลกแห่งความเป็นจริง (เช่น ลูกค้า ผลิตภัณฑ์ คำสั่งซื้อ) ได้โดยตรง ทำให้วัตถุประสงค์ของโค้ดชัดเจนยิ่งขึ้น 

คลาสคือสิ่งที่ใช้รวมข้อมูล (data) และเมธอด (operation) เข้าไว้ด้วยกัน ทำให้ไม่มีการกระจัดกระจายของข้อมูลและความรับผิดชอบ  ในเมื่อแต่ละคลาสถูกสร้างขึ้นเพื่อทำงานเฉพาะอย่าง นั่นหมายความว่า เราสามารถพัฒนาหรือ debug โปรแกรมได้อย่างตรงจุด

การใช้งานคลาสเป็นการเขียนโปรแกรมเชิงวัตถุก็จริง แต่เราสามารถใส่การเขียนโปรแกรมเชิงโครงสร้าง (ซึ่งไม่ทำลายแนวคิดเชิงวัตถุ) ไว้ใน method ของคลาส นั่นหมายความว่าเราสามารถนำความรู้ความเข้าใจหรือแม้กระทั่งโค้ด ที่เคยใช้ในการเขียนโปรแกรมเชิงโครงสร้างมาใช้ในคลาสได้โดยตรง

ในการตั้งชื่อให้กับส่วนต่างๆ ของคลาส (เช่น field, method) จะใช้ชื่อที่มีความหมายตรงไปตรงมา ทำให้ไม่ต้องใช้เวลานานในการทำความเข้าใจโค้ด

ในโปรแกรมที่ออกแบบ class diagram มาเป็นอย่างดีนั้น จะสามารถลดความซับซ้อนและความซ้ำซ้อนลงไปได้เป็นอย่างมาก

เมื่อพัฒนาโปรแกรมไปสักระยะหนึ่ง เราจะพบว่าการการเปลี่ยนแปลงของโค้ดในคลาสจะน้อยลงเรื่อยๆ ถ้าไม่จำเป็นต้องแก้ไขโค้ด เราจะพบว่าเราไม่จำเป็นต้องเปิดไฟล์สำหรับคลาสนั้นๆ เลย ทำให้เราสามารถเก็บโค้ดนั้นลงในไลบรารี่ ซึ่งทำให้ไม่ถูกแก้ไขโดยไม่ตั้งใจ เช่นในระบบตะกร้าสินค้า เราสามารถสร้างคลาสสำหรับ Customer, Product, Order, Payment โดยเพ่งตวามสนใจไปในคลาสที่ได้รับผิดชอบให้เขียนโค้ดเท่านั้น
