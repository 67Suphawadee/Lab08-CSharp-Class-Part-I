## คำถาม

1. Field ในภาษา C# คืออะไร


*คำตอบ*

ตัวแปรที่ถูกประกาศภายในคลาสหรือสตรัคเจอร์ (structure)

2. Field มีหน้าที่อะไรใน class 


*คำตอบ*

เก็บข้อมูลหรือสถานะ (state)ของอ็อบเจ็กต์ที่สร้างจากคลาสนั้น ๆ

3. การประกาศ Field ทำได้อย่างไร ให้อธิบายรูปแบบการประกาศ field ในคลาส


*คำตอบ*

ฟิลด์ (Field) คือตัวแปรที่ถูกประกาศภายในคลาสเพื่อเก็บข้อมูล
ฟิลด์สามารถมีระดับการเข้าถึงที่แตกต่างกัน เช่น public, private, protected, internal
ฟิลด์สามารถกำหนดค่าเริ่มต้นได้ในขณะที่ประกาศ หรือผ่าน constructor
ฟิลด์สามารถถูกควบคุมการเข้าถึงได้ผ่าน properties ที่ใช้ get และ set

4. Field และ Variable แตกต่างกันอย่างไร


Field คือ ตัวแปรที่ประกาศภายในคลาสหรือสตรัคเจอร์และใช้เก็บสถานะของอ็อบเจ็กต์หรือคลาสนั้น ๆ ฟิลด์สามารถมีระดับการเข้าถึงที่สามารถควบคุมได้
Variable คือ ตัวแปรที่ประกาศภายในเมธอดหรือบล็อกโค้ดที่ใช้เก็บข้อมูลในขอบเขตเฉพาะของเมธอดหรือบล็อกนั้น ๆ และจะมีอายุการใช้งานเฉพาะในขอบเขตนั้น ๆ เท่านั้น







5. เมื่อไหร่ควรใช้ Field และเมื่อไหร่ควรใช้ Variable

*คำตอบ*

ควรใช้ Field:
เมื่อข้อมูลต้องเก็บไว้ใน อ็อบเจ็กต์ หรือ คลาส และต้องมีอายุการใช้งานตลอดชีวิตของอ็อบเจ็กต์
เมื่อข้อมูลต้อง ใช้ร่วมกันในหลาย ๆ เมธอด ภายในคลาส
เมื่อข้อมูลมีความสำคัญต่อ สถานะของอ็อบเจ็กต์ (object state)

ควรใช้ Variable:
เมื่อข้อมูลมีอายุการใช้งานแค่ในขอบเขตของ เมธอด หรือ ฟังก์ชัน เท่านั้น
เมื่อข้อมูลไม่จำเป็นต้องเก็บไว้หลังจากการทำงานเสร็จสิ้น
เมื่อข้อมูลเป็น ชั่วคราว และไม่ต้องการให้ผูกพันกับอ็อบเจ็กต์หรือคลาส
